```python
# Get the current directory
current_directory = os.getcwd()

# Define the path to the data folder
data_folder = os.path.join(current_directory, 'data')

# List all files in the data folder
files = os.listdir(data_folder)

# data_2019 = pd.read_csv("/kaggle/input/new-york-city-airbnb-open-data/AB_NYC_2019.csv")


# Read each CSV file in the data folder
for file in files:
    if file.endswith('listings1.csv'):
        print("df1")
        file_path1 = os.path.join(data_folder, file)
        df1 = pd.read_csv(file_path1)
    elif file.endswith('listings2.csv'):
        print("df2")
        file_path2 = os.path.join(data_folder, file)
        df2 = pd.read_csv(file_path2)
    elif file.endswith("listings2021.csv"):
        print("df2021")
        file_path3 = os.path.join(data_folder, file)
        df2021 = pd.read_csv(file_path3)
    elif file.endswith("listings2022.csv"):
        print("df2022")
        file_path4 = os.path.join(data_folder, file)
        df2022 = pd.read_csv(file_path4)
    elif file.endswith("AirBnbNYC_Data.xlsx"):
        print("df2008_2015")
        file_path5 = os.path.join(data_folder, file)
        df2008_2015 = pd.read_excel(file_path5)
    elif file.endswith("AB_NYC_2019.csv"):
        print("df2019")
        file_path6 = os.path.join(data_folder, file)
        df2019 = pd.read_csv(file_path6)
    elif file.endswith("AB_US_2020.csv"):
        print("df2020")
        file_path7 = os.path.join(data_folder, file)
        df2020 = pd.read_csv(file_path7)
    elif file.endswith("Airbnb_NYC_2023_11.csv"):
        print("df2023")
        file_path8 = os.path.join(data_folder, file)
        df2023 = pd.read_csv(file_path8)
        
        
        
        # Process the DataFrame df as needed
```
